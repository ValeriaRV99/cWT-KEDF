```python
Sine_matrices = np.load('/home/vr371/cWT-KEDF/Phases/DATA/Sine_matrices.npy')
sine_Btin = np.asarray(Sine_matrices)[:30,0]
sine_fcc = np.asarray(Sine_matrices)[:30,1]
sine_bcc = np.asarray(Sine_matrices)[:30,2]
sine_8cd = np.asarray(Sine_matrices)[:30,3]
sine_7cd = np.asarray(Sine_matrices)[:30,4]
Matrix = np.vstack((sine_Btin, sine_fcc, sine_bcc, sine_8cd, sine_7cd))

Sine_gradient = np.load('/home/vr371/cWT-KEDF/Phases/DATA/Sine_gradient.npy')
type(Sine_gradient)
dx_Btin, dy_Btin, dz_Btin = Sine_gradient[:,0,0,:,:], Sine_gradient[:,1,0,:,:], Sine_gradient[:,2,0,:,:]
dx_fcc, dy_fcc, dz_fcc = Sine_gradient[:,0,1,:,:], Sine_gradient[:,1,1,:,:], Sine_gradient[:,2,1,:,:]
dx_bcc, dy_bcc, dz_bcc = Sine_gradient[:,0,2,:,:], Sine_gradient[:,1,2,:,:], Sine_gradient[:,2,2,:,:]
dx_8cd, dy_8cd, dz_8cd = Sine_gradient[:,0,3,:,:], Sine_gradient[:,1,3,:,:], Sine_gradient[:,2,3,:,:]
dx_7cd, dy_7cd, dz_7cd = Sine_gradient[:,0,4,:,:], Sine_gradient[:,1,4,:,:], Sine_gradient[:,2,4,:,:]
Gradient = np.vstack((dx_Btin, dy_Btin, dz_Btin, dx_fcc, dy_fcc, dz_fcc, dx_bcc, dy_bcc, dz_bcc, sine_8cd, dx_7cd, dy_7cd, dz_7cd))
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.kernel_ridge import KernelRidge #as modelKR
from sklearn.linear_model import Ridge #as modelR
from sklearn.linear_model import Lasso
from dscribe.descriptors import CoulombMatrix, SineMatrix, EwaldSumMatrix, SOAP
from sklearn.tree import DecisionTreeRegressor
model = KernelRidge()

X = Matrix.reshape(150,8,8)
y = np.asarray(Energy).reshape(150)
energy = []
for j, e in enumerate(y):
        energy_matrix = np.zeros((len(X[1])*len(X[1])))
        energy_matrix[0] = e
        energy.append(energy_matrix)
X = X.reshape(-1, len(X[1])*len(X[1]))
y = np.asarray(energy)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
E_pred = model.predict(Matrix.reshape(-1,8*8))
```
