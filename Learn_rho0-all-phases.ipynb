#### Porpose of this code
This code builds supervised machine learning models to predict $\rho_{0}$ based on the Ewald matrix (or Coulomb matrix for periodic systems). Specifically, we tested three regression models - Kernel Ridge, Ridge, and Lasso - as well as two neural networks - Decision tree and MLPregression - using the data we generated earlier.

#### Load the data

```python
Sine_matrices = np.load('/DATA/Sine_matrices.npy')
Ewal_matrices = np.load('/DATA/Ewal_matrices.npy')
Soap_matrix = np.load('/DATA/Soap_matrix.npy', allow_pickle=True)

rho0_Btin = np.load("/DATA/Si-B-tin/DATA2/rho0.npy") #
rho0_fcc = np.load("/DATA/Si-fcc/DATA2/rho0.npy") #
rho0_bcc = np.load("/DATA/Si-bcc/DATA2/rho0.npy") #
rho0_8cd = np.load("/DATA/Si8-CD/DATA2/rho0.npy") #
rho0_7cd = np.load("/DATA/Si7-CD/DATA2/rho0.npy") 
```

All the data will be compare with KS calculations, so we load the KS total energy of the system

```python
l = np.linspace(0.8, 1.4, 30)
VOL = []
KS_TE = []
for i, X in enumerate(l):
    with open('.../Si-B-tin/DATA2/Si'+str(i)+'.out') as D:
        k = [match for match in D if "Total energy" in match]
    KS_te = str(k).split()[6]
    KS_TE.append(float(KS_te)) 
    ions = Ions.from_ase(crystal('Si',[(0,0,0)],spacegroup=141,cellpar = [4.81,4.81,2.65, 90,90,90]))
    cell = ions.get_cell()
    ions.set_cell(cell * X, scale_atoms=True)
    vol=ions.get_volume()
    VOL.append(vol)
```

#### Arange the data

```python
esm_Btin = np.asarray(Ewal_matrices)[:30,0]
esm_fcc = np.asarray(Ewal_matrices)[:30,1]
esm_bcc = np.asarray(Ewal_matrices)[:30,2]
esm_8cd = np.asarray(Ewal_matrices)[:30,3]
esm_7cd = np.asarray(Ewal_matrices)[:30,4]
```

We concatenate all the matrices and rho0 in one array

```python
rho0 = np.hstack((rho0_Btin, rho0_fcc, rho0_bcc, rho0_8cd.reshape(30), rho0_7cd.reshape(30)))

Matrix = np.vstack((esm_Btin, esm_fcc, esm_bcc, esm_8cd, esm_7cd))
```

To treat rho0 we create a vector of the size of the matrix, in the first spot we store the value rho0 fof the respective Ewal matrix

```python
X = Matrix.reshape(150,8,8)
y = np.asarray(rho0)
rh0 = []
for j, r0 in enumerate(y):
        rho0_matrix = np.zeros((len(X[1])*len(X[1])))
        rho0_matrix[0] = r0
        rh0.append(rho0_matrix)
X = X.reshape(-1, len(X[1])*len(X[1]))
y = np.asarray(rh0)
```

#### Learning using Kernel Ridge regression
In the following cells we apply ML implementing Kernel ridge regression to learn rho0 parameter
$$
X = Ewal\ sum \ matrix
$$
$$
y = \rho_{0}
$$

To find the best value alpha for the regression we build the function alpha_min

```python
def alpha_min(x0, *args):
    model, X_train, y_train, X_test, y_test = args
    model.set_params(alpha=float(x0))
    model.fit(X_train,y_train)
    return 1 - np.abs(model.score(X_test,y_test))
```

We compare Kernel ridge, Ridge and Lasso regression

```python
from sklearn.model_selection import train_test_split 
from sklearn.kernel_ridge import KernelRidge #as modelKR
from sklearn.linear_model import Ridge #as modelR
from sklearn.linear_model import Lasso #as modelL

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
model_kr = KernelRidge()
model_r = Ridge()
model_L = Lasso()
minn_kr = minimize(alpha_min, 0.001, args = (model_kr, X_train, y_train, X_test, y_test), method='Nelder-Mead', bounds=[[0, 1]], options={'ftol' : 1e-8, 'disp':True})
minn_r = minimize(alpha_min, 0.001, args = (model_r, X_train, y_train, X_test, y_test), method='Nelder-Mead', bounds=[[0, 1]], options={'ftol' : 1e-8, 'disp':True})
minn_L = minimize(alpha_min, 0.001, args = (model_L, X_train, y_train, X_test, y_test), method='Nelder-Mead', bounds=[[0, 1]], options={'ftol' : 1e-8, 'disp':True})

print(minn_kr.x)
print(minn_r.x)
print(minn_L.x)
```

```python
model_kr.set_params(alpha=0.001)
model_kr.fit(X_train, y_train)
print(model_kr.score(X_test,y_test))
model_r.set_params(alpha=0.00105)
model_r.fit(X_train, y_train)
print(model_r.score(X_test,y_test))
model_L.set_params(alpha=0.001)
model_L.fit(X_train, y_train)
print(model_L.score(X_test,y_test))
```

#### Applaying the model
Once the model was set up we predict the rho0 values using any phase of Si and building the Ewal matrix

```python
from dscribe.descriptors import CoulombMatrix, SineMatrix, EwaldSumMatrix, SOAP

r = np.linspace(0.8, 1.4, 30)
VOLUME = []
rh0_kr =[]
rh0_r = []
rh0_L =[]
for i, d in enumerate(r):
    si_Btin = Ions.from_ase(crystal('Si',[(0,0,0)],spacegroup=141,cellpar = [4.81,4.81,2.65, 90,90,90]))
    cell = si_Btin.get_cell()
    si_Btin.set_cell(cell * d, scale_atoms=True)
    vol=si_Btin.get_volume()    
    esm = EwaldSumMatrix(n_atoms_max=8)
    X_pol = esm.create(si_Btin)
    y_kr = model_kr.predict(X_pol.reshape(-1, len(X_pol)))
    y_r = model_r.predict(X_pol.reshape(-1, len(X_pol)))
    y_L = model_L.predict(X_pol.reshape(-1, len(X_pol)))
    rh0_kr.append(y_kr[0,0])
    rh0_r.append(y_r[0,0])
    rh0_L.append(y_L[0,0])
    VOLUME.append(vol)
```

We compare the real rho0 and the learn rho0

```python
plt.plot(np.asarray(VOLUME)*0.529177**3/4,np.asarray(rho0[0:30]),'b*--', label='cWT')
plt.plot(np.asarray(VOLUME)*0.529177**3/4, np.asarray(rh0_kr),'r*--', label='KR')
plt.plot(np.asarray(VOLUME)*0.529177**3/4, np.asarray(rh0_r),'g*--', label='R')
plt.plot(np.asarray(VOLUME)*0.529177**3/4, np.asarray(rh0_L),'k*--', label='Lasso')
plt.xlabel('Volume ($\AA^{3}$)')
plt.ylabel('rho0')
plt.title('Si-\u03B2$_{tin}$')
plt.legend(['cWT','KR', 'R', 'Lasso'])
```

Now we perform an OFDFT calculation implementing the learn rho0 into the cWT KEDF

```python
l = np.linspace(0.8, 1.4, 30)
rho00 = np.asarray(rh0_kr)
XC = Functional(type='XC',name='LDA')
HARTREE = Functional(type='HARTREE')

ENERGY = []
VOLUME = []
for i in np.arange(0, 30, 1):
        KE = Functional(type='KEDF',name='WT', rho0=rho00[i])
#         ions = Ions.from_ase(bulk('Si', 'diamond', a=5.43, cubic=True))
        ions = Ions.from_ase(crystal('Si',[(0,0,0)],spacegroup=141,cellpar = [4.81,4.81,2.65, 90,90,90]))
        cell = ions.get_cell()
        ions.set_charges(4)
        ions.set_cell(cell * l[i], scale_atoms=True)
        nr = ecut2nr(ecut=25, lattice=ions.cell)
        grid = DirectGrid(lattice=ions.cell, nr=nr)
        rho_ini = DirectField(grid=grid)
        rho_ini[:] = ions.get_ncharges()/ions.cell.volume
        PSEUDO = LocalPseudo(grid = grid, ions=ions, PP_list=PP_list, rcut=10)
        evaluator = TotalFunctional(KE=KE, XC=XC, HARTREE=HARTREE, PSEUDO=PSEUDO)
        
        optimization_options = {'econv' : 1e-5*ions.nat}
        opt = Optimization(EnergyEvaluator=evaluator, optimization_options = optimization_options, 
                           optimization_method = 'TN')
        rho = opt.optimize_rho(guess_rho=rho_ini)

        energy = evaluator.Energy(rho=rho, ions=ions)
#         Force = GetForces(ions=ions, rho=rho, EnergyEvaluator=evaluator)
        vol=ions.get_volume()
        ENERGY.append(energy)
        VOLUME.append(vol)
```

```python
plt.plot(np.asarray(VOLUME)*0.529177**3/4,(np.asarray(ENERGY))*27.211407/4)
plt.plot(np.asarray(VOLUME)*0.529177**3/4, np.asarray(KS_TE)*27.211407/8)
plt.xlabel('Volume ($\AA^{3}$)')
plt.ylabel('Energy (eV/atom)')
plt.title('KernelRidge Si-\u03B2$_{tin}$')
plt.legend(['cWT','KS'])
```

#### Learning using a neural network Decission tree
In the following two cells we apply ML implementing neural network to learn rho0 parameter
X = Ewal matrix composed from all the phases
y = rho0 parameter

```python
X_Btin = Matrix.reshape(120,8,8)
y_Btin = np.asarray(rho0)
rho0 = []
for j, r0 in enumerate(y_Btin):
        rho0_matrix = np.zeros((8))
        rho0_matrix[0] = r0
        rho0.append(rho0_matrix)
X = X_Btin.reshape(-1, len(X_Btin[1])*len(X_Btin[1]))
y = np.asarray(rho0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
tree = DecisionTreeRegressor(max_depth=25)

mlp = make_pipeline(
    MLPRegressor(
        hidden_layer_sizes=(30, 15),
        learning_rate_init=0.01,
        early_stopping=True,
        random_state=0,
    ),
)

tree.fit(X,y)
print(tree.score(X, y))

mlp.fit(X, y)
print(mlp.score(X, y))
```

```python
r = np.linspace(0.8, 1.4, 30)
VOLUME = []
rh0_tree = []
rh0_mlp = []
for i, d in enumerate(r):
    si = Ions.from_ase(crystal('Si',[(0,0,0)],spacegroup=141,cellpar = [4.81,4.81,2.65, 90,90,90]))
    cell = si.get_cell()
    si.set_cell(cell * d, scale_atoms=True)
#     traj.append(si)
    vol = si.get_volume()
    
    esm = EwaldSumMatrix(n_atoms_max=8)
    X_pol = esm.create(si, n_jobs=2)

    y_tree = tree.predict(X_pol.reshape(-1, len(X_Btin[1])*len(X_Btin[1])))
    y_mlp = mlp.predict(X_pol.reshape(-1, len(X_Btin[1])*len(X_Btin[1])))

    rh0_tree.append(y_tree[0,0])
    rh0_mlp.append(y_mlp[0,0])
    VOLUME.append(vol)
```
